# **Data Analysis**

In the data analysis, we shall answer the folllowing questions:


*  What is the most popular hour for returning cars? 
*  What station is the most popular?
*  At the most popular picking hour?
*  What postal code is the most popular for picking up Blue cars?
*  Does the most popular station belong to that postal code?
*   At the most popular picking hour?
*   Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars? 
*   Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.





```python
# Most popular hour of the day for picking up a shared electric car in Paris In April 2018
city1 = car[car['City'] == 'Paris']
hour1= city1.groupby(['Public name', 'hour'])[['Bluecar counter']].sum()
hour1.head(1)
```

```python
#Most popular hour for returning cars 
hour1['bluecar_usage'] = hour1['Bluecar counter'].diff()
usage = hour1[hour1['bluecar_usage'] < 0]
usage1=usage.groupby('hour')[['bluecar_usage']].sum().sort_values('bluecar_usage', ascending=True)
usage1.head(1)
```

```python
#Most popular station
## finding the difference in number of cars available at the station using the diff() method

car['bluecar_usage'] = car['Bluecar counter'].diff()

# creating column for utilib usage
car['utilib_usage'] = car['Utilib counter'].diff()

# creating column for utilib 14 usage
car['utilib_14_usage'] = car['Utilib 1.4 counter'].diff()
car.head(1)
```

```python
#Most popular picking hour
df2= df[(df['City'] == 'Paris') & (car['bluecar_usage'] < 0)]
df2.head()
car1 = car.groupby(['hour'])[['bluecar_usage']].sum().sort_values('bluecar_usage', ascending=True)
car1.head(1)
```

```python
#Most Postal codes with most Blue cars pickups
car[(car['Kind'] == 'STATION') & (car['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
```

```python
#Most popular station belonging to the most popular postal code at most popular picking hour
car[(car['Kind'] == 'STATION') & (car['Status'] == 'ok') & (car['hour'] == 4)].groupby('Postal code')['Postal code'].count().sort_values(ascending= False).head(1)
```

```python
# Create date feature
paris_df['date'] = paris_df[paris_df.columns[18:21]].apply(lambda x: '/'.join(x.astype(str)), axis=1)
# Create time feature
paris_df['time'] = paris_df[paris_df.columns[21:23]].apply(lambda x: ":".join(x.astype(str)), axis = 1)
#Create datetime feature
paris_df['datetime'] = paris_df[paris_df.columns[23:]].apply(lambda x: " ".join(x.astype(str)), axis = 1)

```
